### NumPy

[NumPy](https://numpy.org): The package which is not a built-in module but is treated like one by most of the community.
To start with Numpy, we must start by understanding the fundamental storage format in NumPy.

[Arrays](https://numpy.org/doc/stable/user/absolute_beginners.html#what-is-an-array). All numpy functions are made to be used on arrays. We are supposed to use numpy arrays instead of python lists for two primary reasons.

1. NumPy arrays model our mathematics better because NumPy was made with numeric computation in mind.
2. It is faster than python lists

The second point however is a double edged sword.
A python list does not need to be homogenous, a numpy array has to be homogenous.
Homogeneity in this case refers to having the same data type of elements inside the array. They need to belong to one type like integers or complex numbers; more accurately to their [datatype equivalents](https://numpy.org/doc/stable/reference/arrays.scalars.html#built-in-scalar-types).

## Curve Fitting: The 🌠 of this show

Curve fitting is special, so I shall give it the special treatment. Instead of talking about just set of inputs and outputs. I want to show two cases of curve fitting.

### Example 1

```python
# importing stuff I need
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
x_data = np.array([0.        , 0.15789474, 0.31578947, 0.47368421, 0.63157895,
       0.78947368, 0.94736842, 1.10526316, 1.26315789, 1.42105263,
       1.57894737, 1.73684211, 1.89473684, 2.05263158, 2.21052632,
       2.36842105, 2.52631579, 2.68421053, 2.84210526, 3.        ])
y_data = np.array([  2.95258285,   2.49719803,  -2.1984975 ,  -4.88744346,
        -7.41326345,  -8.44574157, -10.01878504, -13.83743553,
       -12.91548145, -15.41149046, -14.93516299, -13.42514157,
       -14.12110495, -17.6412464 , -16.1275509 , -16.11533771,
       -15.66076021, -13.48938865, -11.33918701, -11.70467566])
```

```python
# Lets try to plot it to see how it looks like
plt.scatter(x_data,y_data)
plt.show()
```

Some additional information, the experiment that I have collected this data from has an equation that is of the form:
$$
f(x)=a(x-b)^2 + c
$$

Now that we have this, let us start trying to find the best fit curve.

```python
# We will do this in 3 different steps

"""
Step 1:
Defining the function
"""

def model_f(x, a, b, c):
    return a*(x-b)**2 + c
```

```python
"""
Step 2:
Applying the function
"""
popt, pcov = curve_fit(model_f, x_data, y_data)
```

```python
"""
Step 3:
Getting the coefficients we were looking for
"""
a, b, c = popt
print(f"a:{a}\nb:{b}\nc:{c}")
```

```python
"""
Step 4:
PLotting the new curve alongside the old scatter to see how good of a result we got
"""
x_model = np.linspace(np.min(x_data), np.max(x_data), 1000)
y_model = model_f(x_model, a, b, c)
plt.scatter(x_data,y_data)
plt.plot(x_model,y_model, color='r')
plt.show()
```

This is our first curve that we fit:
$$
f(x) = a(x-b)^2 + c \:\:\:\: \text{where a = 4.922, b = 2.004, c = -16.04}
$$

## A new day a new curve

The equation for a string motion is $y(t)=A \cos{(\omega t  +\phi)}$. We went to the lab and collected some data and now we want to figure out what the values of $A$, $\omega$ and $\phi$ might be.

```python
t_data = np.array([ 0.   ,  0.34482759,  0.68965517,  1.03448276,  1.37931034,
        1.72413793,  2.06896552,  2.4137931 ,  2.75862069,  3.10344828,
        3.44827586,  3.79310345,  4.13793103,  4.48275862,  4.82758621,
        5.17241379,  5.51724138,  5.86206897,  6.20689655,  6.55172414,
        6.89655172,  7.24137931,  7.5862069 ,  7.93103448,  8.27586207,
        8.62068966,  8.96551724,  9.31034483,  9.65517241, 10.        ])
y_data = np.array([ 4.3303953 ,  1.61137995, -2.15418696, -3.90137249, -1.67259042,
        2.16884383,  3.86635998,  1.85194506, -1.8489224 , -3.96560495,
       -2.13385255,  1.59425817,  4.06145238,  1.89300594, -1.76870297,
       -4.26791226, -2.46874133,  1.37019912,  4.24945607,  2.27038039,
       -1.50299303, -3.46774049, -2.50845488,  1.20022052,  3.81633703,
        2.91511556, -1.24569189, -3.72716214, -2.54549857,  0.87262548])
```

```python
"""
Step 0:
Plot it the data once
"""
plt.scatter(t_data, y_data)
plt.show()
```

```python
"""
Step 1:
Define the function
"""
def model_f(t, A, omega, phi):
    return A*np.cos(omega*t + phi)
```

```python
"""
Step 2:
Use the curve_fit function
"""
popt, pcov = curve_fit(model_f, t_data, y_data)
```

```python
"""
Step 3:
Finding the coefficients we desire
"""
A, omega, phi = popt
print(f"A: {A}\nOmega: {omega}\nPhi: {phi}")
```

```python
"""
Step 4:
Plotting the new curve
"""
plt.scatter(t_data, y_data)
model_x = np.linspace(np.min(t_data), np.max(t_data), 1000)
model_y = model_f(model_x, A, omega, phi)
plt.plot(model_x, model_y)

plt.show()
```

No one would blame you if you lost all faith in SciPy after seeing this.
Did I go wrong somewhere while making this?

The real answer is:

SciPy is not magic. Sometimes you must give it some initial guesses on the value of these coefficients.

Let's try that.

```python
plt.plot(t_data,y_data,'o--')
plt.show()
```

Looking at this, what could be some good initial guesses to the values.

$A = 4$ roughly, $\omega = 2 \pi f$ and $f = 1/T$, $T=2$ very roughly from this picture and talking about $\phi$ it looks roughly equal to 0

```python
# Lets try to curve fit again
popt, pcov = curve_fit(model_f, t_data, y_data, p0=(4, np.pi, 0))
# This time we are passing in an additional argument called p0
# Its a tuple containing our initial guesses of A, omega and phi

A, omega, phi = popt
print(f"A: {A}\nOmega: {omega}\nPhi: {phi}")

# Lets plot this
plt.scatter(t_data, y_data)
model_x = np.linspace(np.min(t_data), np.max(t_data), 1000)
model_y = model_f(model_x, A, omega, phi)
plt.plot(model_x, model_y)

plt.show()
```
